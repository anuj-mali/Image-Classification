## Dependencies

```python
import torch
from torch import nn
from torchvision.datasets import Imagenette
from torchvision.transforms import v2, ToTensor
from torch.utils.data import DataLoader

from matplotlib import pyplot as plt
import numpy as np
```

## Device Agnostic Code

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

## Datasets and DataLoaders

### Datasets

```python
transform = v2.Compose([
    ToTensor(),
    v2.Resize((320,320))
])
```

```python
train_dataset = Imagenette(root='./', split='train', size='320px', transform=transform)
test_dataset = Imagenette(root='./', split='val', size='320px', transform=transform)
```

### DataLoaders

```python
import os

train_dataloader = DataLoader(dataset=train_dataset,
                             batch_size=32,
                             pin_memory=True,
                             shuffle=True,
                              num_workers=1)
test_dataloader = DataLoader(dataset=test_dataset,
                             batch_size=32,
                             pin_memory=True,
                             shuffle=False,
                              num_workers=1)
```

## Create AlexNet model

**Red is not error. The cause is tqdm**

```python
class CNNBlock(nn.Module):
    def __init__(self, in_channels, out_channels, **kwargs):
        super().__init__()
        self.conv_layer = nn.Conv2d(in_channels=in_channels,
                                   out_channels=out_channels,
                                   **kwargs)
        self.max_pool = nn.MaxPool2d(stride=2, kernel_size=3)
        self.relu = nn.ReLU()
        self.local_response = nn.LocalResponseNorm(size=5, k=2)

    def forward(self, x):
        return self.max_pool(self.local_response(self.relu(self.conv_layer(x))))
```

```python
class AlexNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv_block_1 = CNNBlock(in_channels=3,
                                     out_channels=96,
                                     padding='valid',
                                     kernel_size=11,
                                     stride=4)
        self.conv_block_2 = CNNBlock(in_channels=96,
                                    out_channels=256,
                                    padding='same',
                                    kernel_size=5)
        self.conv_layer_1 = nn.Conv2d(in_channels=256,
                                     out_channels=384,
                                     kernel_size=3,
                                     padding='same')
        self.conv_layer_2 = nn.Conv2d(in_channels=384,
                                     out_channels=384,
                                     kernel_size=3,
                                     padding='same')
        self.conv_layer_3 = nn.Conv2d(in_channels=384,
                                     out_channels=256,
                                     kernel_size=3,
                                     padding='same')
        self.max_pool = nn.MaxPool2d(kernel_size=3,
                                    stride=2)
        self.relu = nn.ReLU()
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=256*8*8, out_features=4096),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(in_features=4096, out_features=4096),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(in_features=4096, out_features=10)
        )

        self._init_weights()

    def _init_weights(self):
        for module in self.modules():
            if isinstance(module, nn.Conv2d) or isinstance(module, nn.Linear):
                nn.init.kaiming_uniform_(module.weight)

    def forward(self, x):
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        x = self.relu(self.conv_layer_1(x))
        x = self.relu(self.conv_layer_2(x))
        x = self.relu(self.conv_layer_3(x))
        x = self.max_pool(x)
        x = self.classifier(x)
        return x
```

```python
from torchinfo import summary

summary(AlexNet(), input_size=[128, 3, 320, 320])
```

```python
model = AlexNet().to(device)
```

```python
from torch.optim import Adam

loss_fn = nn.CrossEntropyLoss()
optimizer = Adam(params=model.parameters(), lr=0.0001)
```

```python
from tqdm.auto import tqdm

def train_step(model: torch.nn.Module,
              dataloader: torch.utils.data.DataLoader,
              loss_fn: torch.nn.Module,
              optimizer: torch.optim.Optimizer,
              device: str):
    mean_loss = 0
    mean_acc = 0
    model.train()
    for X,y in tqdm(dataloader):
        X,y = X.to(device), y.to(device)
        y_logits = model(X)
        loss = loss_fn(y_logits, y)
        mean_loss += loss.item()
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        # Calculate and accumulate accuracy metric across all batches
        y_pred_class = torch.argmax(torch.softmax(y_logits, dim=1), dim=1)
        mean_acc += (y_pred_class == y).sum().item()/len(y_logits)

    return mean_loss/len(dataloader), mean_acc/len(dataloader)
```

```python
def test_step(model: torch.nn.Module,
              dataloader: torch.utils.data.DataLoader,
              loss_fn: torch.nn.Module,
              device: str):
    mean_loss = 0
    mean_acc = 0
    model.eval()
    with torch.inference_mode():
        for X,y in tqdm(dataloader):
            X,y = X.to(device), y.to(device)
            y_logits = model(X)
            loss = loss_fn(y_logits, y)
            mean_loss += loss.item()
            
            # Calculate and accumulate accuracy metric across all batches
            y_pred_class = torch.argmax(torch.softmax(y_logits, dim=1), dim=1)
            mean_acc += (y_pred_class == y).sum().item()/len(y_logits)

    return mean_loss/len(dataloader), mean_acc/len(dataloader)
```

```python
for epoch in range(10):
    train_loss, train_acc = train_step(model, train_dataloader, loss_fn, optimizer, device)
    test_loss, test_acc = test_step(model, test_dataloader, loss_fn, device)

    print(f'Epoch: {epoch} | Loss: {train_loss: .4f} | Acc: {train_acc: .3f} | Loss: {test_loss: .4f} | Acc: {test_acc: .3f}')
```

```python
import torchvision.models as models

# Define model
model = models.resnet18(weights='DEFAULT')
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, len(train_dataset.classes))

model.to(device)

optimizer = Adam(model.parameters(), lr=0.001)

for epoch in range(10):
    train_loss, train_acc = train_step(model, train_dataloader, loss_fn, optimizer, device)
    test_loss, test_acc = test_step(model, test_dataloader, loss_fn, device)

    print(f'Epoch: {epoch} | Loss: {train_loss: .4f} | Acc: {train_acc: .3f} | Loss: {test_loss: .4f} | Acc: {test_acc: .3f}')
```
